```python
import pandas as pd

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

Read in data

```python
data = pd.read_csv("apps_all_background.txt")
data.head()
```

PCA with 95% explained variance

```python

x = data.drop(["TARGET", "SK_ID_CURR"], axis = 1)

scaler = StandardScaler()
x_std = scaler.fit_transform(x)

var_explained = 0.95

pca = PCA(n_components = var_explained)
principal_components = pca.fit_transform(x_std)

explained_var = pca.explained_variance_ratio_

df_pca = pd.DataFrame({
    "Principal Component": [f"PC{i+1}" for i in range(len(explained_var))],
    "Explained Variance Ratio": explained_var
})

print(df_pca)
```
